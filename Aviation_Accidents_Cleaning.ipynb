## Data Cleaning

### Filtering aircrafts and events

We want to filter the dataset to include aircraft that the client is interested in an analysis of:
- inspect relevant columns
- figure out any reasonable imputations
- filter the dataset

```python
# Standardize date and filter to 1983–2023 (40-year window), keep professional airplane builds and accidents
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df = df[df['Event.Date'].notna()].copy()
df['Event.Year'] = df['Event.Date'].dt.year

# Investigation type: accidents only
df = df[df['Investigation.Type'].str.strip().str.upper() == 'ACCIDENT']

# Professional builds only (exclude amateur/homebuilt where marked Yes/True)
df['Amateur.Built'] = df['Amateur.Built'].astype(str).str.strip().str.upper()
df = df[~df['Amateur.Built'].isin(['YES', 'Y', 'TRUE', '1'])].copy()

# Airplane category only (exclude helicopters, balloons, etc.)
if 'Aircraft.Category' in df.columns:
    df = df[df['Aircraft.Category'].astype(str).str.upper() == 'AIRPLANE']

# Year filter
df = df[(df['Event.Year'] >= 1983) & (df['Event.Year'] <= 2023)].copy()
df.shape
```

### Cleaning and constructing Key Measurables

Injuries and robustness to destruction are a key interest point for the client. Clean and impute relevant columns and then create derived fields that best quantifies what the client wishes to track. **Use commenting or markdown to explain any cleaning assumptions as well as any derived columns you create.**

**Construct metric for fatal/serious injuries**

*Hint:* Estimate the total number of passengers on each flight. The likelihood of serious / fatal injury can be estimated as a fraction from this.

```python
# Construct severe injury metrics
for col in ['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

df['Estimated.Total.Occupants'] = (
    df[['Total.Fatal.Injuries','Total.Serious.Injuries','Total.Minor.Injuries','Total.Uninjured']]
      .sum(axis=1, min_count=1)
)

df['Severe.Injury.Count'] = (
    df[['Total.Fatal.Injuries','Total.Serious.Injuries']].sum(axis=1, min_count=1)
)

df['Severe.Injury.Rate'] = (
    df['Severe.Injury.Count'] / df['Estimated.Total.Occupants']
)

# Flag any accident with at least one serious or fatal injury
df['Severe.Injury.Flag'] = (df['Severe.Injury.Count'].fillna(0) > 0)

df[['Estimated.Total.Occupants','Severe.Injury.Count','Severe.Injury.Rate','Severe.Injury.Flag']].head()
```

**Aircraft.Damage**
- identify and execute any cleaning tasks
- construct a derived column tracking whether an aircraft was destroyed or not.

```python
# Clean aircraft damage and derive total destruction flag
if 'Aircraft.damage' in df.columns:
    df['Aircraft.damage'] = df['Aircraft.damage'].astype(str).str.strip().str.title()
    # Normalize common variants
    replacements = {
        'Destroyed (Substantial)': 'Destroyed',
        'Substantial Damage': 'Substantial',
        'Unknown': pd.NA,
        'UNK': pd.NA
    }
    df['Aircraft.damage'] = df['Aircraft.damage'].replace(replacements)
    df['Total.Destruction.Flag'] = df['Aircraft.damage'].eq('Destroyed')
else:
    df['Total.Destruction.Flag'] = False

df['Total.Destruction.Flag'].value_counts(dropna=False)
```

### Investigate the *Make* column
- Identify cleaning tasks here
- List cleaning tasks clearly in markdown
- Execute the cleaning tasks
- For your analysis, keep Makes with a reasonable number (you can put the threshold at 50 though lower could work as well)

```python
# Clean and standardize Make
df['Make'] = df['Make'].astype(str).str.strip().str.upper().replace({'': pd.NA, 'NAN': pd.NA})
make_counts = df['Make'].value_counts(dropna=True)
display(make_counts.head(20))

# Keep makes with at least a threshold number of accidents for robust stats (e.g., >= 50)
MAKE_MIN_N = 50
df['Make.Count'] = df['Make'].map(make_counts)
df = df[df['Make.Count'] >= MAKE_MIN_N].copy()
df.drop(columns=['Make.Count'], inplace=True, errors='ignore')
print("Remaining rows after make filter:", df.shape[0])
```

### Inspect Model column
- Get rid of any NaNs.
- Inspect the column and counts for each model/make. Are model labels unique to each make?
- If not, create a derived column that is a unique identifier for a given plane type.

```python
# Clean Model and create a unique type identifier
df['Model'] = df['Model'].astype(str).str.strip().str.upper().replace({'': pd.NA, 'NAN': pd.NA})
# Drop rows where Model is missing
df = df[df['Model'].notna()].copy()

# Create a combined Make-Model identifier
df['Make_Model'] = (df['Make'].fillna('') + ' ' + df['Model'].fillna('')).str.strip()
df['Make_Model'] = df['Make_Model'].replace({'': pd.NA})

# Check uniqueness of model labels within a make
counts_by_make_model = df.groupby(['Make','Model']).size().sort_values(ascending=False).head(10)
display(counts_by_make_model)

df[['Make','Model','Make_Model']].head()
```

### Cleaning other columns
- there are other columns containing data that might be related to the outcome of an accident. We list a few here:
- Engine.Type
- Weather.Condition
- Number.of.Engines
- Purpose.of.flight
- Broad.phase.of.flight

Inspect and identify potential cleaning tasks in each of the above columns. Execute those cleaning tasks. 

**Note**: You do not necessarily need to impute or drop NaNs here.

```python
# Standardize selected categorical columns
def clean_cat(series):
    return (series.astype(str)
                  .str.strip()
                  .str.replace(r'\s+', ' ', regex=True)
                  .str.title()
                  .replace({'Unk': pd.NA, 'Unknown': pd.NA, 'Nan': pd.NA, '': pd.NA}))

for col in ['Engine.Type','Weather.Condition','Purpose.of.flight','Broad.phase.of.flight']:
    if col in df.columns:
        df[col] = clean_cat(df[col])
        
# Numeric engines
if 'Number.of.Engines' in df.columns:
    df['Number.of.Engines'] = pd.to_numeric(df['Number.of.Engines'], errors='coerce')

# Show cleaned columns snapshot
cols_to_show = [c for c in ['Engine.Type','Weather.Condition','Number.of.Engines','Purpose.of.flight','Broad.phase.of.flight','Size.Segment'] if c in df.columns]
df[cols_to_show].head()
```

### Column Removal
- inspect the dataframe and drop any columns that have too many NaNs

```python
# Drop columns with too many missing values (e.g., > 70% NaN)
missing_frac = df.isna().mean().sort_values(ascending=False)
to_drop = missing_frac[missing_frac > 0.7].index.tolist()
df = df.drop(columns=to_drop)
print("Dropped columns due to missingness > 70%:", to_drop)
print("Remaining shape:", df.shape)
```
